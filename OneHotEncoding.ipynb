```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('data/COMPLETE_df.pkl',"rb") as f:
    data = pickle.load(f)

data['class'] = pd.Categorical(data['class'])
dataDummies = pd.get_dummies(data['class'], prefix = 'category')
data = pd.concat([data, dataDummies], axis=1)
```
